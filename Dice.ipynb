```python
import numpy as np

def roll_dice():
    return np.random.randint(1, 7)  # 1 to 6

def simulate_rolls(n):
    rolls = np.random.randint(1, 7, size=n)
    return rolls

def show_stats(rolls):
    print("\n🎲 Roll Results:", rolls)
    
    # Count frequency
    unique, counts = np.unique(rolls, return_counts=True)
    
    print("\n📊 Frequency:")
    for u, c in zip(unique, counts):
        print(f"{u}: {c} times")

def main():
    print("===== Dice Simulator =====")
    
    while True:
        choice = input("\n1. Roll once\n2. Roll multiple times\n3. Exit\nEnter choice: ")
        
        if choice == '1':
            print("You rolled:", roll_dice())
        
        elif choice == '2':
            n = int(input("How many times to roll? "))
            rolls = simulate_rolls(n)
            show_stats(rolls)
        
        elif choice == '3':
            print("Goodbye 👋")
            break
        
        else:
            print("Invalid choice!")

main()
```
